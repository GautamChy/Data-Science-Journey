Standardization(Z-Score)

```python
import numpy as np
mean = np.mean(total_bill)
std = np.std(total_bill)
```

```python
mean
```

```python
std
```

```python
mean - std , mean + std
```

```python
data = []
for i in total_bill:
  z_score = (i - mean)/std
  data.append(z_score)
```

```python
print(data)
```

```python
max(data)
```

```python
min(data)
```

```python
max(data) - min(data)
```

```python
sns.histplot(data)
```

# Normalization(Min-Max Scaler)

```python
import seaborn as sns
```

```python
df = sns.load_dataset('taxis')
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
```

```python
# fit use garnu ko reason 'distance', 'fare', 'tip', 'total' yo vitra ko data bujhxa or read garchha.
min_max_scaler.fit(df[['distance', 'fare', 'tip', 'total']])
```

```python
# transform use garnu ko reason 'change the entire data by applying formula'
min_max_scaler.transform(df[['distance', 'fare', 'tip', 'total']])
```

```python
df2 = pd.DataFrame(min_max_scaler.transform(df[['distance', 'fare', 'tip', 'total']]),columns=['distance', 'fare', 'tip', 'total'])
```

```python
df2.head()
```

```python
max(df2['distance']), min(df2['distance'])
```
