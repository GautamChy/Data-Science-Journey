```python
## Feature Engineering
## .str.split('/') → splits the string at /, turning "24/03/2019" into ["24", "03", "2019"]
df['Day'] = df['Date_of_Journey'].str.split('/').str[0]  #.str[0] → takes the first element → "24", It creates a new column Day with the day value
df['Month'] = df['Date_of_Journey'].str.split('/').str[1] # Takes the second element → "03", Creates a Month column
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]  # Takes the third element → "2019" , Creates a Year column
```

```python
df.head()
```

```python
df.info()
```

* These lines are converting the new columns (Day, Month, Year) from strings (text) into integers (numbers).

* df['Date_of_Journey'].str.split('/')

* the result was text like "24", "03", "2019" — this is string not integer

# To convert string into integer

* df['Day'] = df['Day'].astype(int)
     * Converting string into integer "24" → 24

* df['Month'] = df['Month'].astype(int)
     * Converting string into integer "03" → 3

* df['Year'] = df['Year'].astype(int)
     * Converting string into integer "2019" → 2019

```python
# Converting string into integer
df['Day'] = df['Day'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
## Drop the Date_of_Journey column
df.drop('Date_of_Journey', axis=1, inplace=True)
```
