# Tecchniques to handle imbalanced dataset
* Upsamping : It means ghataidinney
* Downsamping: It means ghataidiney

 # Upsampling:
   * Upsampling means increasing the number of samples in the minority class.
  
  Goal:
  * Make classes more balanced.

  Mothods:
  * Duplicate existing minority samples
* Generate synthetic data (like SMOTE)

Example:
*  Before:

* Fraud = 50
* No Fraud = 950

* After upsampling:

* Fraud = 950
* No Fraud = 950

Advantages:
* Keeps all original data
* Helps model learn minority patterns better

Disadvantages:
* Can cause overfitting if duplicates are repeated too much

# Downsampling
* Downsampling means reducing the number of samples in the majority class.

Example

* Before:

* Fraud = 50
* No Fraud = 950

After downsampling:

* Fraud = 50
* No Fraud = 50

Advantages:
* Faster training
* Reduces dominance of majority class

Disadvantages:
* Loses potentially useful information




# Simple analogy
* Imagine a classroom vote:

* 95 students prefer tea
* 5 prefer coffee

To make opinions balanced:

* Upsampling → ask the 5 coffee students to speak multiple times
* Downsampling → randomly choose only 5 tea students to participate



These methods are especially common in:

* Fraud detection
* Medical diagnosis
* Spam detection
* Rare event prediction

# Upsamping
* Upsampling means increasing the number of samples in the minority class.

```python
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
df_minority.shape, df_majority.shape
```

```python
from sklearn.utils import resample
# replace=True → allows duplicates → required for upsampling
# n_samples=len(df_majority) → expands minority class to match majority size
# Upsampling → replace=True
df_minority_upsampled = resample(df_minority, replace=True, n_samples =  len(df_majority)) # resample vitra (df_minority, replace=True, n_samples =  len(df_majority)) haalney.
                                                                                          # replace = True means increase gardiney i.e Upsampling ma jailey replace = True hunxa.
                                                                                          # n_samples =  len(df_majority)) it means length increase gardiney.
```

# Downsampling
* Downsampling means reducing the number of samples in the majority class.

```python
from sklearn.utils import resample
# replace=False means sampling without replacement
# n_samples=len(df_minority) is likely smaller than the majority class
# Downsampling → replace=False
df_majority_downsampling = resample(df_majority, replace=False, n_samples =  len(df_minority)) # replace = True means decrease gardiney i.e Upsampling ma jailey replace = False hunxa.
```
