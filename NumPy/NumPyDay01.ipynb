**NumPy Array**

- Array is a data structure that stores a sequence of elements of the same data type. Arrays can be one-dimensional (a simple sequence), two-dimensional (rows and columns), or multi-dimensional.

- List can store different types and numpy array stores only one data type (all integers or all floats).
- Numpy array is faster as compared to python list.
- Basic operations like adding, removing or iterating is done in list while advanced math operations like mean, sum or matrix multiplication can be performed in numpy array.
- List uses more memory while array uses less memory and is more efficient for large data.-

NumPy arrays are better than Python lists because they are built using C, making them faster and more efficient. They also support vectorized operations, allowing you to perform math on entire arrays at once, which makes handling large amounts of data much easier.

**Array Attributes**

Array attributes are simple properties of NumPy arrays that tell us basic details about the array, like:

- Its size (number of elements).
- Its shape (how many rows and columns it has).
- The type of data it holds.
- The number of dimensions it has (1D, 2D, 3D, etc.).
- These attributes help us understand and work with the array easily.

**Attributes and Their Meanings:**

- ndim: Number of dimensions (e.g., 1 for a 1D array).
- shape: Size of the array in each dimension (e.g., length for 1D).
- size: Total number of elements in the array.
- dtype: Type of data stored in the array.
- itemsize: Size of each element in bytes.
- nbytes: Total memory used by the array (itemsize × size).

```python
# create a NumPy array
numbers = np.array([2, 4, 6, 8, 10, 12, 14])

# give the array dimension
print('Number of dimensions:', numbers.ndim)

# get shape of the array
print('Shape:', numbers.shape)
```

```python
numbers1 = np.array([2,4,6,8,10,'a'])
print(numbers1)
print("Data Types of First Array:", numbers1.dtype)
```

```python
# create an array of int type
int_numbers = np.array([2,4,6,8,10,12,13])
# check data types of number
print("Data type :",int_numbers.dtype )
```

```python
# create an array of float type
float_numbers = np.array([2,4,6,8,10,12,13])
# check data type of float array
print("Data Types of Third Array:", float_numbers.dtype)
```

We use size attribute to find the total number of elements in the array.

```python
first = np.array([1,3,5,7,8,9,12])
second = np.array([1,2,4,6,10])

print('Size of first:', first.size)
print('Size of second:', second.size)
```

We use itemsize attribute to find the size(in bytes) of each element in the array.

```python
# create a default 1-D array of integers
array1 = np.array([6, 7, 8, 9, 10, 11])

# create a 1-D array of 32-bit integers
array2 = np.array([6, 7, 8, 9, 10, 11], dtype=np.int16)

# create a 1-D array of floats
array3 = np.array([1.2,2.4,5.6,7.8], dtype=np.float128)

print(array1)
print(array2)
print(array3)

print(array1.itemsize)
print(array2.itemsize)
print(array3.itemsize)
```

nbytes finds the total number of bytes consumed by elements in the array.

```python
print(array1.nbytes)
print(array2.nbytes)
print(array3.nbytes)
```

```python
array4 = np.array([1,2,3,4,5], dtype = np.float32)
print(array4)
print(array4.itemsize)
print(array4.nbytes)
```

```python
a = np.array([10, 12, 15, 21])
b = np.array([15, 12, 10, 21])

# use of less()
result = np.less(a, b) # <
print(result)

# use of less_equal()
result = np.less_equal(a, b) # <=
print(result)

# use of greater()
result = np.greater(a, b) # >
print(result)

# use of greater_equal()
result = np.greater_equal(a, b) # >=
print(result)

# use of equal()
result = np.equal(a, b) # ==
print(result)

# use of not_equal()
result = np.not_equal(a, b) # !=
print(result)
```
