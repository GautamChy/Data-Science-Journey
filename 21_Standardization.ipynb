```python
# Display the total_bill column as a list
list(df['total_bill'])
```

```python
# assign the total_bill column to a variable
total_bill = list(df['total_bill'])
```

```python
import numpy as  np # import the numpy library
mean = np.mean(total_bill) # calculate the mean of the total_bill column
std = np.std(total_bill) # calculate the standard deviation of the total_bill column
```

```python
# Display the mean and standard deviation
mean, std
```

```python
normalized_data = [] # create an empty list called normalized_data
for i in total_bill: # loop through the total_bill column
  z_score = (i - mean) / std # calculate the z_score i.e standardization(z_score) formula.
  normalized_data.append(z_score) # append the z_score to the data
```

```python
# Display the normalized_data_set
normalized_data
```

```python
 # Dsiplay the maximum value in the normalized_data_set
max(normalized_data)
```

```python
# Dsiplay the minimum value in the normalized_data_set
min(normalized_data)
```

# Standardization

# using API to work directly

```python
# import the StandScaler class from the sklearn.preprocessing, Search on google www.StandardScaler.skearn
from sklearn.preprocessing import StandardScaler
```

```python
# create an instance of the StandardScale class
scaler = StandardScaler()
```

```python
scaler # When we click on StandardScaler there are show Not fitted, it means its empty not fit anything.
```

```python
# Display the first 5 rows of the dataset
df.head()
```

```python
# fit the StandardScaler to the total_bill and tip column. Now when we click on StandardScaler there will be shown fitted,it means data fitted i.e 'total_bill, tip'
scaler.fit(df[['total_bill','tip']])
```

```python
# transform the total_bill and tip columns
scaler.transform(df[['total_bill','tip']])
```

```python
# Convert into dataFrame
import pandas as pd
df = pd.DataFrame(scaler.transform(df[['total_bill','tip']]), columns = ['total_bill_standard','tip_standard'])
```

```python
# Display the maximum value in the total_bill column
max(df['total_bill_standard'])
```

```python
# Display the minimum in the total_bill column
min(df['total_bill_standard'])
```

```python
# Display the maximum value in the tip column
max(df['tip_standard'])
```

```python
# Display the minimum value in the tip column
min(df['tip_standard'])
```
